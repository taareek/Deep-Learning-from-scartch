# Building Network using Pytorch

```python
from torch import nn
```

```python
class Network(nn.Module):
    def __init__(self):
        super().__init__()
        #Inputs to hidden layers transformation
        self.hidden = nn.Linear(784, 256)
        #Output layers, 10- units for each digits
        self.output = nn.Linear(256,10)
        
        #define sigmoid and softmax output
        sigmoid = nn.Sigmoid()
        softmax = nn.Softmax(dim =1)
        
    def forward(self, x):
        #pass the input tenors through each of our operations
        x = self.hidden(x)
        x = self.output(x)
        x = self.sigmoid(x)
        x = softmax(x)
        
        return x
    
```

```python
# Create the network and look at it's text representation
model = Network()
model
```

```python
import torch.nn.functional as F

class Network2(nn.Module):
    def __init__(self):
        super().__init__()
        # Inputs to hidden layer linear transformation
        self.hidden = nn.Linear(784, 256)
        # Output layer, 10 units - one for each digit
        self.output = nn.Linear(256, 10)
        
    def forward(self, x):
        # Hidden layer with sigmoid activation
        x = F.sigmoid(self.hidden(x))
        # Output layer with softmax activation
        x = F.softmax(self.output(x), dim=1)
        
        return x
```

```python
class MyNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        #input hidden layers in linear transformation
        self.hdn1 = nn.Linear(784, 128) #First hidden layer
        self.hdn2 = nn.Linear(128, 64) #Second hidden layer
        self.output_f = nn.Linear(64, 10) #Output layer
        #Define ReLu and softmax function
        relu = nn.ReLU()
        softmax = nn.Softmax(dim = 1)
        
    def forward(self,x):
        x = self.hdn1(x)
        x = self.relu(x)
        x = self.hdn2(x)
        x = self.relu(x)
        x = self.output_f(x)
        x = self.softmax(x)
        
        return x

        
```
